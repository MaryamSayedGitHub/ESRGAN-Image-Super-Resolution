```python
# inference.py
import torch
import cv2
import numpy as np
from RRDBNet_arch import RRDBNet

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_path = '/content/RRDB_PSNR_x4.pth'
model = RRDBNet(3, 3, 64, 23, gc=32)
# model.load_state_dict(torch.load(model_path), strict=True)
torch.load(model_path, weights_only=True)
model.load_state_dict(torch.load(model_path, weights_only=True), strict=True)


model.eval()
model = model.to(device)

def enhance_image(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    img = img * 1.0 / 255
    img = torch.from_numpy(np.transpose(img[:, :, [2, 1, 0]], (2, 0, 1))).float()
    img = img.unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(img).data.squeeze().float().cpu().clamp_(0, 1).numpy()
    output = np.transpose(output[[2, 1, 0], :, :], (1, 2, 0))
    output = (output * 255.0).round().astype(np.uint8)
    return output
```

```python
from matplotlib import pyplot as plt
import cv2
# Enhance the image
output = enhance_image("/content/comic.png")

# Display the result
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Enhanced Image')
plt.show()
```

```python
cv2.imwrite("enhanced_image.png", output)
```
